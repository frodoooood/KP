```python
def generate_and_save_image(model, epoch, test_input):
    prediction = model(test_input, training=False)
    fig = plt.figure(figsize=(4, 4))

    for i in range(prediction.shape[0]):
        plt.subplot(4, 4, i + 1)
        pred = (np.array(prediction[i]) * 255.)
        pred = pred.astype(int)
        plt.imshow(pred)
        plt.axis('off')
    if epoch % 10 == 0:
        plt.savefig('image_at_epoch_{:04d}.png'.format(epoch))
    plt.show()
    
```

```python
desc = K.models.load_model('/content/drive/MyDrive/descriminator_dna/descriminator_h10_dna.h5')
gen = K.models.load_model('/content/drive/MyDrive/generator_dna/generator_h10_dna.h5')
START_EPOCH = 0
```

```python
cross_entropy = K.losses.BinaryCrossentropy(from_logits=True)


def loss_gen(fake_out, fake_img, real_img):
    x = K.backend.reshape(fake_img, shape=(BATCH_SIZE, 64*64*3))
    # x = x.astype('int')
    decoder = K.backend.reshape(real_img, shape=(BATCH_SIZE, 64*64*3))
    delta = cross_entropy(decoder,x)
    # print(delta)   # decoder = dexoder.astype('int')
    # delta = 64*64*3*tf.keras.losses.BinaryCrossentropy(x, decoder)
    return (cross_entropy(tf.ones_like(fake_out), fake_out) + delta)

def loss_desc(fake_out, real_out):
    real_loss = cross_entropy(tf.ones_like(real_out), real_out)
    fake_loss = cross_entropy(tf.zeros_like(fake_out), fake_out)
    return real_loss + fake_loss


gen_optimazer = K.optimizers.Adam(1e-4)
desc_optimazer = K.optimizers.Adam(1e-4)


EPOCH = 10000
num_examples_to_generate = 16
# seed = tf.random.normal([num_examples_to_generate, latent_dim])
# seed = np.random.normal(0,1, size=((num_examples_to_generate,)+latent_dim))
r = random.randint(0,1500)
seed = data[r:r+num_examples_to_generate]
```

```python
@tf.function
def train_step(image_iput, latent_batch, epoch):
  # noise = K.backend.random_normal([BATCH_SIZE, latent_dim])
  # noise = np.random.normal(0,1, size=((BATCH_SIZE,)+latent_dim))
  with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
      gen_image = gen(latent_batch, training=True)
      real_out = desc(image_iput, training=True)
      fake_out = desc(gen_image, training=True)
      gen_losses = loss_gen(fake_out, gen_image, image_iput)
      disc_losses = loss_desc(fake_out, real_out)
  grad_of_gen = gen_tape.gradient(gen_losses, gen.trainable_variables)
  grad_of_desc = disc_tape.gradient(disc_losses, desc.trainable_variables)
  gen_optimazer.apply_gradients(zip(grad_of_gen, gen.trainable_variables))
  desc_optimazer.apply_gradients(zip(grad_of_desc, desc.trainable_variables))

  return gen_losses, disc_losses
```

```python
gen.trainable = True
desc.trainable = True
```

```python
def train(dataset, latent_dim, start_epoch, epochs):
    MAX_PRIN_LABEL = 10
    th = BUFFER_SIZE // (BATCH_SIZE * MAX_PRIN_LABEL)

    for epoch in range(start_epoch, epochs):
        print(f"{epoch+1}/{epochs}:", end='')
        start = time.time()
        n = 0
        gen_loss_epoch = 0
        disc_loss_epoch = 0
        for image_batch, latent_batch in zip(dataset, latent_dim):
            # dna_inp = data_fake[n*BATCH_SIZE:n*BATCH_SIZE+BATCH_SIZE]
            # dna_inp = tf.reshape(dna_inp, (BATCH_SIZE, 1, 1, 200))
            gen_loss, disc_loss = train_step(image_batch, latent_batch, epoch)
            gen_loss_epoch += K.backend.mean(gen_loss)
            disc_loss_epoch += K.backend.mean(disc_loss)
            if (n % th == 0): print('=', end='')
            n += 1

        print(': Ошибка генератора: ' + str(gen_loss_epoch / n) + " Ошибка дескрименатора: " + str(disc_loss_epoch / n), end='')
        print(' Время эпохи {} составляет {} секунд'.format(epoch + 1, time.time() - start))
        
        display.clear_output(wait=True)
        generate_and_save_image(gen, epoch + 1, seed)
        gen.save(f'/content/drive/MyDrive/generator_dna/generator_h10_dna.h5')
        desc.save(f'/content/drive/MyDrive/descriminator_dna/descriminator_h10_dna.h5')

    display.clear_output(wait=True)
    generate_and_save_image(gen, epoch, seed)
```

```python
train(train_dataset, latent_dataset, START_EPOCH, EPOCH)
```
